```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def analyze_qos(vec_file, sca_file):
    # Baca file vektor
    vec_data = pd.read_csv(vec_file, delimiter='\t')
    
    # Baca file scalar
    sca_data = pd.read_csv(sca_file, delimiter='\t')
    
    # Analisis Throughput
    throughput = vec_data[vec_data['name'].str.contains('throughput')]
    plt.figure(figsize=(10,5))
    plt.plot(throughput['time'], throughput['value'])
    plt.title('Throughput')
    plt.show()
    
    # Ringkasan Statistik
    print("Throughput Statistics:")
    print(throughput['value'].describe())
    
    # Analisis Delay
    delay = vec_data[vec_data['name'].str.contains('packetDelay')]
    print("\nDelay Statistics:")
    print(delay['value'].describe())

# Gunakan fungsi
analyze_qos('results/ScenarioA-0.vec', 'results/ScenarioA-0.sca')
```

```python
def read_vec_file(filename):
    data = []
    with open(filename, 'r') as file:
        for line in file:
            if line.startswith('vector'):
                parts = line.split()
                vector_id = parts[1]
                vector_name = parts[2]
            elif not line.startswith('#'):
                try:
                    time, value = map(float, line.strip().split('\t'))
                    data.append({
                        'vector_id': vector_id,
                        'vector_name' : vector_name,
                        'time' : time,
                        'value' : value
                    })
                except:
                    pass
    return pd.DataFrame(data)

def read_sca_file(filename):
    data = {}
    with open(filename, 'r') as file:
        for line in file:
            if line.startswith('scalar'):
                parts = line.split()
                module = parts[1]
                metric = parts[2]
                value = float(parts[3])

                if module not in data:
                    data[module] = {}
                data[module][metric] = value
    return pd.DataFrame.from_dict(data, orient='index')

# Analisis QoS
def analyze_qos(vec_file, sca_file):
    # Baca file
    vec_data = read_vec_file(vec_file)
    sca_data = read_sca_file(sca_file)
    
    # Analisis Throughput
    throughput = vec_data[vec_data['vector_name'].str.contains('throughput')]
    
    # Visualisasi Throughput
    plt.figure(figsize=(12,6))
    plt.subplot(2,2,1)
    plt.plot(throughput['time'], throughput['value'])
    plt.title('Throughput Over Time')
    plt.xlabel('Waktu (s)')
    plt.ylabel('Throughput (bps)')
    
    # Analisis Delay
    delay = vec_data[vec_data['vector_name'].str.contains('delay')]
    plt.subplot(2,2,2)
    plt.plot(delay['time'], delay['value'])
    plt.title('Packet Delay Over Time')
    plt.xlabel('Waktu (s)')
    plt.ylabel('Delay (s)')
    
    # Statistik Deskriptif
    print("Statistik Throughput:")
    print(throughput['value'].describe())
    
    print("\nStatistik Delay:")
    print(delay['value'].describe())
    
    # Uji Distribusi
    plt.subplot(2,2,3)
    stats.probplot(throughput['value'], dist="norm", plot=plt)
    plt.title('Q-Q Plot Throughput')
    
    plt.subplot(2,2,4)
    stats.probplot(delay['value'], dist="norm", plot=plt)
    plt.title('Q-Q Plot Delay')
    
    plt.tight_layout()
    plt.show()
    
    # Ringkasan Scalar
    print("\nRingkasan Scalar:")
    print(sca_data)
    
    # Analisis Energi
    energy_consumption = vec_data[vec_data['vector_name'].str.contains('energy')]
    plt.figure(figsize=(10,5))
    plt.plot(energy_consumption['time'], energy_consumption['value'])
    plt.title('Konsumsi Energi')
    plt.xlabel('Waktu (s)')
    plt.ylabel('Energi (J)')
    plt.show()

# Tambahan: Korelasi antar metrik
def analyze_correlation(vec_file):
    vec_data = read_vec_file(vec_file)
    
    # Pivot data untuk korelasi
    pivot_data = vec_data.pivot(index='time', columns='vector_name', values='value')
    
    # Hitung korelasi
    correlation_matrix = pivot_data.corr()
    
    plt.figure(figsize=(10,8))
    plt.imshow(correlation_matrix, cmap='coolwarm')
    plt.colorbar()
    plt.title('Matriks Korelasi Metrik QoS')
    plt.xticks(range(len(correlation_matrix.columns)), correlation_matrix.columns, rotation=45)
    plt.yticks(range(len(correlation_matrix.columns)), correlation_matrix.columns)
    plt.show()
    
    print("Matriks Korelasi:")
    print(correlation_matrix)
    
```

```python
# Jalankan analisis
# Ganti dengan path file .vec dan .sca Anda
analyze_qos('results/ScenarioA-0.vec', 'results/ScenarioA-0.sca')

# Analisis korelasi
# Ganti dengan path file .vec Anda
analyze_correlation('results/ScenarioA-0.vec')
```
